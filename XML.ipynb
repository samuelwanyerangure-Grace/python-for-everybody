```python
##eXtensible Markup language
#it has tags and attributes serialization ands deserialization
#doesn't care much about whitespace
#the main tag then the chidren
#when we put an attribute they can be many but we canm only have one text node when we use attribute
import xml.etree.ElementTree as ET
#this is only a string by now
data= ''' 
<person>
    <name>Chuck</name>
    <phone type='intl'>
        +1 734 303 4456
    </phone>
    <email hide='yes'/>
</person>
'''
#when we ET.frostring(data) it parses it to become a xml
#works as beautiful soup
tree = ET.fromstring(data)
print(f'Element {tree.find('name').text }')
print(f'Attribute {tree.find('email').get('hide')} ')
```

```python
import xml.etree.ElementTree as ET

inp='''
<stuff>
    <users> 
        <user x='2'>
            <id>001</id>
            <name>Chuck</name>
        </user>

        <user x='7'>
            <id>003</id>
            <name>Samuel</name>
        </user>

    </users>
</stuff>    
'''
Parsd=ET.fromstring(inp)
lst=Parsd.findall('users/user')
print(f'All the users are here {len(lst)}')
#getting into it them
for item in lst:
    print(item.find('name').text)
    print(item.find('id').text)
    print(item.get('x'))
```
