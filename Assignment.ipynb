```python
#mow onto something more lively
#as it uses less space
#used in huge datasets
count=0
total=0
#when true continue doing the thing your doing
while True:
    num=input('enter a number')
    if num=='Stp':
        break
    try:
        num=int(num)
        count=count+1
        total=total+num
    except ValueError:
        print('Invalid input')
        continue
print('total=',total, 'count=',count, 'average=',total/count)
```
