```python
#using a shortcut give by DR. Chuck
fhand=open('mbox-short.txt')
counts=dict()
for line in fhand:
    line=line.rstrip()
    words=line.split()
    for word in words:
        counts[word]=counts.get(word,0)+1
#print(counts)
#simpler way doing all this: formulae
srt=(sorted( [ (v,k) for k,v in counts.items() ],reverse=True ) )
#print(srt)
for v,k in srt[:10]:
    print(k,v)
```
