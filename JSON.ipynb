```python
#VERY MORE JSON WILL BE FOUND COMPARED TO XML
#javascript Object Notation
#here python processing json
#JSON IS A DOUBLE QOUTE CLUB ONLY HAHAAHAHA
import json
data= '''{
'name' : 'Chuck' , 
'phone' : { 'type' : 'int', 'number' : '+1 734 303 4456'},

'email' : {'hide' : 'yes'}
}'''
#it parses it and returns it as a dictionary
#because this is a curly brace info is a dictionary
info= json.loads(data)
print('Name',info['name'])
print(f'Hide: {info['email']['hide']}')
```
