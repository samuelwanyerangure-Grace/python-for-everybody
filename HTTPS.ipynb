```python
import socket
#AF_INET says that it should use the IPV4
#SOCK_STREAM says that we use TCP=it is reliable and our data will not get lost
#UDP connection is used in video games etc
mysock= socket.socket(socket.AF_INET, socket.SOCK_STREAM)
#use ipv4 and TCP thats what the code above says
mysock.connect( ('data.pr4e.org',80))
#this above is the actual connection
#we decided to use http when using https we would have used TLS

#GET we want everything displayed
#the others like post,delete and put are reserved not easy for us to use
cmd = 'GET http://data.pr4e.org/romeo.txt HTTP/1.0\r\n\r\n'.encode()
mysock.send(cmd)
while True:
    data=mysock.recv(512)  
    if len(data)<1:
        break
    print(data.decode())
mysock.close()
```

```python
import socket
mysock=socket.socket(socket.AF_INET, socket.SOCK_STREAM)
mysock.connect(('data.pr4e.org',80))
#already established a connection here
#so what you want to do
cmd= 'GET http://data.pr4e.org/romeo-full.txt HTTP/1.0\r\n\r\n'.encode()
mysock.send(cmd)
#the request is sent over the internet
while True:
    data=mysock.recv(512)
    if len(data)<1:
        break
    print(data.decode())
mysock.close()
```

```python
#only intrested with the data right hence why we remove the need of the passport
#bt its risky because of the man in the middle attacks
import urllib
from bs4 import BeautifulSoup
import ssl

#ignore ssl certificate errors
#this is what l was referring to as a 'passport' we ignore if the server has no passport
#at first what we are doing this context() is a new bodyguard and its very strict.
ctx = ssl.create_default_context()
#we now change some of the rules
#here you don't care on the host name
ctx.check_hostname=False
#here u also don't verify with third parties
ctx.verify_mode = ssl.CERT_NONE

url=input('enter url')
#we are ensuring that the server assumes that its is communicating to a chrome browser hahahahaha
headers= {'User-agent': 'Mozilla/5.0(Windows NT 10.0; Win64; x64)'}
req=urllib.request.Request(url, headers=headers)
html=urllib.request.urlopen(req,context=ctx).read()
soup=BeautifulSoup(html, 'html.parser')

#retriving all anchor tags
tags=soup('a')
for tag in tags:
    print(tag.get('href', None))  
```
