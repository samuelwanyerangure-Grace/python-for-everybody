```python
#INHERITANCE BUILDING ON CLASSES
#Makes a new class based on existing class
class Person:
    def __init__(self,name):
        self.name=name

    def greet(self):
        print(f'Hello, l am {self.name}')

class Student(Person):
    def __init__(self,name,marks):
        super().__init__(name) #inherit inititalization from Person
        self.marks=marks

    def grade(self):
        if self.marks>=70:
            return 'A'
        elif self.marks>=50:
            return 'B'
        else:
            return 'C'
s=Student('Sam',34)
s.greet()
print(s.marks)
s.grade()
```

```python
class BankAccount:
    bank_name='Equity Group'
    def __init__(self,owner,balance):
        self.owner=owner
        self.balance=balance
    def deposit(self,deposit):
        return print(f'You have a {self.balance + deposit} bank balance')

    def withdraw(self,withdraw):
        return print(f'You have a balance of {self.balance-withdraw}')

p1=BankAccount('Samuel',2000)
p1.withdraw(10000)
print(p1.balance)
```

```python
class BankAccount:
    #class attribute
    bank_name='Equity Group'
    #using two instance
    def __init__(self,owner,balance):
        self.owner=owner
        self.balance=balance
    def deposit(self,deposit):
        self.balance+=deposit 
        print(f'Hello {self.owner}. Your balance sh {self.balance} at {self.bank_name}')

    def withdraw(self,withdraw):
        self.balance -= withdraw
        print(f'Hello {self.owner}. Your balance is {self.balance}')

p1=BankAccount('Samuel',1000)
p1.withdraw(500)
p2=BankAccount('Peniel',50000)
p2.deposit(5000)
```
