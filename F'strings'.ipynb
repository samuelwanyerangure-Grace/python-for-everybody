```python
#F''strings onbe of the cleanest tools in python
#print(f'....') means a formatted string
#lets u insert variables directly inside a string
name='pal'
print(f'Hello {name}')
#instead of 
print('Hello' + name)
#Anything inside {} gets evaluated(run as Python code)
age=20
print(f'l am {age} years old')
#Expressions this is where it gets really powerful
x=5
y=10
print(f'The sum is {x+y}')
#Real case(Used a lot)
name='Pal'
score=95
print(f'{name} scored {score}% in the test ')

#General structure
#(value:format)
#value(your variable)  :(format thisn variable)   format(rules you give)
#Always put the f before quotes
pi=3.14159
print(f'Pi rounded: {pi:.2f}')

#ALIGMENT AND SPACING HAHAHAHA (:<10)
name='Max'
print(f'{name:<10}') #< this means u left align. 10 is the total wdth

print(f'{name:>10}')

print(f'{name:^10}')

#COMBINED WTH LABELS
name='Pal'
score=95

print(f'Name: {name:<20} Score: {score}')
```

```python
#Playing with a dictionary
name=['pal', 'Sam', 'dada', 'susan']
score=[32,32,21,80]
grade=['C','C','D','A']
for name,score,grade in zip(name, score ,grade):
    print(f'{name:<20} {score} {grade}')

#what they do when the data is less
name=['pal', 'Sam', 'dada']
score=[32,32]
for pair in zip(name,score):
    print(pair)
#its stops at the shortest list
#turning zip into a list
pairs=list(zip(name,score))
print(f'{pairs}')
```
